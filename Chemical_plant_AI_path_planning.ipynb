```python
import heapq
```

```python
start = (1, 1)
goal = (18, 18)
```

```python
# ---------------------------------
# Robot Movement Directions
# ---------------------------------

moves = [
    (-1, 0),   # Up
    (1, 0),    # Down
    (0, -1),   # Left
    (0, 1),    # Right

    (-1, -1),  # Upper Left
    (-1, 1),   # Upper Right
    (1, -1),   # Lower Left
    (1, 1)     # Lower Right
]
```

```python
# ---------------------------------
# Movement Cost Function
# ---------------------------------

def get_cell_cost(cell_value):
    """
    Returns the movement cost of each cell.
    """

    if cell_value == 0:      # Empty Floor
        return 1

    elif cell_value == 3:    # Hazard Zone
        return 4

    elif cell_value in [1, 2, 6]:   # Wall, Tank, Machinery
        return float('inf')

    else:
        return 1
```

```python
# ---------------------------------
# Movement Cost
# ---------------------------------

import math

def movement_cost(current, next_pos):

    dx = abs(current[0] - next_pos[0])
    dy = abs(current[1] - next_pos[1])

    # Diagonal movement
    if dx == 1 and dy == 1:
        return math.sqrt(2)

    # Horizontal / Vertical movement
    return 1
```

```python
import heapq
import math

# ---------------------------------
# Heuristic Function
# ---------------------------------

def heuristic(a, b):
    return math.sqrt((a[0] - b[0])**2 + (a[1] - b[1])**2)


# ---------------------------------
# A* Path Planning Algorithm
# ---------------------------------

def astar(plant, start, goal):

    open_set = []

    heapq.heappush(open_set, (0, start))

    came_from = {}

    g_score = {start: 0}

    f_score = {start: heuristic(start, goal)}

    while open_set:

        current = heapq.heappop(open_set)[1]

        if current == goal:

            path = []

            while current in came_from:

                path.append(current)

                current = came_from[current]

            path.append(start)

            return path[::-1]

        for move in moves:

            neighbor = (
                current[0] + move[0],
                current[1] + move[1]
            )

            # Check boundaries
            if not (0 <= neighbor[0] < plant.shape[0] and
                    0 <= neighbor[1] < plant.shape[1]):
                continue

            cell = plant[neighbor]

            # Skip walls, tanks and machinery
            if cell in [1,2,6]:
                continue

            tentative_g = (
                g_score[current]
                + movement_cost(current, neighbor)
                + get_cell_cost(cell)
            )

            if neighbor not in g_score or tentative_g < g_score[neighbor]:

                came_from[neighbor] = current

                g_score[neighbor] = tentative_g

                f_score[neighbor] = (
                    tentative_g +
                    heuristic(neighbor, goal)
                )

                heapq.heappush(
                    open_set,
                    (f_score[neighbor], neighbor)
                )

    return None
```

```python
path = astar(plant, start, goal)

print(path)
```

```python
# ---------------------------------
# Draw the AI Path
# ---------------------------------

# Create a copy so the original plant is not modified
display_plant = plant.copy()

# Draw the path
for position in path:

    row, col = position

    # Don't overwrite robot or goal
    if display_plant[row, col] not in [4, 5]:
        display_plant[row, col] = 7
```

```python
import math

def heuristic(a, b):
    dx = abs(a[0] - b[0])
    dy = abs(a[1] - b[1])

    D = 1
    D2 = math.sqrt(2)

    return D * (dx + dy) + (D2 - 2 * D) * min(dx, dy)
```

```python
path = astar(plant, start, goal)
print(path)
```

```python
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# Make a copy of the original plant
animation_plant = plant.copy()

fig, ax = plt.subplots(figsize=(8,8))

def animate(frame):

    ax.clear()

    current_map = animation_plant.copy()

    # Draw completed path
    for i in range(frame):
        r, c = path[i]
        if current_map[r, c] not in [4,5]:
            current_map[r, c] = 7

    # Current robot position
    r, c = path[frame]
    current_map[r, c] = 4

    ax.imshow(current_map, cmap=cmap)

    ax.set_title("AI Robot Navigation in Chemical Plant")

    ax.set_xticks(range(20))
    ax.set_yticks(range(20))

    ax.grid(color='gray')

ani = FuncAnimation(
    fig,
    animate,
    frames=len(path),
    interval=300,
    repeat=False
)

plt.close()

HTML(ani.to_jshtml())
```
